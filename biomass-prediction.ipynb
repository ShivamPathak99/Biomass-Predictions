**Dataset: Biomass_History.csv**

**Importing essentila libraries**

```python
from fastai.tabular.all import *
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('/kaggle/input/agriculture/Biomass_History.csv')
```

```python
X_train = data[['Latitude', 'Longitude', '2010', '2011', '2012', '2013', '2014', '2015', '2016']]
y_train = data['2017']
```

**Using Random Forest**

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
X_test = data[['Latitude', 'Longitude', '2010', '2011', '2012', '2013', '2014', '2015', '2016']]
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)
```

```python
rf.fit(X_train, y_train)
```

```python
predictions = pd.DataFrame({'Index': data['Index'], 'Predicted Biomass Production': y_pred})
```

```python
print(predictions)
```

```python
rmse = mean_squared_error(data['2017'], y_pred, squared=False)
```

```python
print('RMSE:', rmse)
```

```python
predictions.to_csv('Predicted_Biomass.csv', index=False)
```
